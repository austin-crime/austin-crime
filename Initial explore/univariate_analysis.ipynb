### Takeaways

- crime_type: Has a lot of categories making it difficult to see how the feature is distributed. Having broader categories would be ideal but setting this up would be time consuming. For a first pass through the pipeline we should try to reduce the number of categories possibly by removing categories with less than N observations, where N is decided by someone.
- family_violence: Most crimes do not involve family violence, but many crimes also do not involve violence period.
- occurence_date: Roughly uniform distribution. There may be some seasonality.
- occurence_time: Too many unique values to get any meaningful insights.
- report_date and report_time: Same as occurence_date and occurence_time.
- location_type: Too categories to gain any meaningful insights, but not so many categories that it would be necessary to reduce the data.
- address: Probably not very useful.
- zip_code: Looks like some zip codes do tend to have larger crime counts.
- council_districts: A few council districts have higher crime counts than the others.
- sector and district: There are certain sectors and districts that have low counts of crime. This would need further investigation. Other than those the rest are roughly uniformly distributed.
- latitude and longitude: Roughly normally distributed.
- clearance_date: Roughly uniform distribution.


- clearance_status: It will probably be useful to combine cleared by arrest and cleared by exception into one category. There is an imbalance in the target variable, but for our first pass through the pipeline we will leave it as is and use either recall or precision as our performance metric in modeling. In our second iteration we'll try downsampling the data to have a more balanced target to see if that leads to any improvements.
